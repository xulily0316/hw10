```python
import numpy as np
import pandas as pd
import datetime as dt
import sqlalchemy
from sqlalchemy import create_engine, inspect
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///C:/Users/Chloe/OneDrive/Desktop/Bootcamp2/Challenge10_sqlalchemy/Resources/hawaii.sqlite", echo=False)
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
session = Session(engine)
```

```python
# Participation Analysis
from sqlalchemy import inspect
inspector = inspect(engine)
inspector.get_table_names()
```

```python
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c["type"])
```

```python
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c["type"])
```

```python
most_recent_dt =session.query(Measurement.date).order_by(Measurement.date.desc()).first()
print("Most recent date in database: " , most_recent_dt)

year_ago = dt.date(2017, 8, 23) - dt.timedelta(days = 365)
print("Date one year ago is: ", year_ago)
```

```python
results = session.query(Measurement.prcp , Measurement.date).\
    filter(Measurement.date > '2016-08-23').\
    order_by(Measurement.date).all()
results
```

```python
df_results=pd.DataFrame(results)
df_results=df_results.rename(columns={"prcp":"precipitation"})

df_results.sort_values(by=['date'], ascending=True)
df_results=df_results.dropna()
df_results=df_results.reset_index(drop=True)
```

```python
df_results.plot(x='date',figsize=(20,10),legend=True,linewidth=9.0,grid=True)
plt.xlabel('Date')
plt.ylabel('Inches')

plt.show()
```

```python
#Station Analysis
import sqlalchemy
from sqlalchemy import create_engine, func
station_count = session.query(func.count(Measurement.station)).all()
station_count
```

```python
most_active_stations = session.query( Measurement.station , func.count(Measurement.station)).group_by( Measurement.station ).\
               order_by(func.count(Measurement.station).desc()).all()

most_active_stations
```

```python
temp_stations = session.query(func.min(Measurement.tobs), func.max(Measurement.tobs),\
                      func.avg(Measurement.tobs)).\
                      filter(Measurement.station == 'USC00519281' ).all()


print("The Minimum ,Maximum and Average Temperatures for the most active station is :" , temp_stations)
```

```python
lastest_dt = session.query (Measurement.date).filter(Measurement.station == 'USC00519281').\
          order_by(Measurement.date.desc()).first()
print("Latest date for station 'USC00519281' is: " , lastest_dt )
```

```python
results_temp =  session.query(Measurement.tobs).\
                          filter(Measurement.station == 'USC00519281' ).\
                          filter(Measurement.date >= '2017,8,23').all()

df_temp=pd.DataFrame(results_temp)
```

```python
hist = df_temp['tobs'].hist(bins=12)
plt.title("Temperatures observed for station 'USC00519281' in the last 12 months" ,
           size = 15)
plt.xlabel('Temperature')
plt.ylabel('Frequency', labelpad=20, weight='bold', size=12)
labels = ['tobs']
plt.legend(labels)

plt.show()
```
